# Nutzung BeautifulSoup

```python
r = requests.get("https://www.immowelt.de/liste/wuerzburg/wohnungen/mieten?d=true&sd=DESC&sf=RELEVANCE&sp=1")
```

```python
r.encoding = 'utf-8'
```

```python
# print Seite als HTML
html = r.text
html
```

```python
print(html)
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
soup.title
```

```python
# gibt nur den ersten zurück
soup.title.string
```

```python
soup.title.string.parent
```

```python
h2 = []
for tag in soup.findAll('h2'):
  h2.append(tag.string)
  #print(tag.string)
```

```python
h2
```

```python
'''for tag in soup.findAll('div', {'class':"KeyFacts-efbce"}):
  print(tag)'''
```

```python
price = []
for tag in soup.findAll('div', {'data-test':"price"}):
  price.append(tag.string)
  #print(tag.string)
```

```python
qm = []
for tag in soup.findAll('div', {'data-test':"area"}):
  qm.append(tag.string)
  #print(tag.string)
```

```python
zimmer = []
for tag in soup.findAll('div', {'data-test':"rooms"}):
  zimmer.append(tag.string)
  #print(tag.string)
```

```python
'''for tag in soup.findAll('div', {'class':"IconFact-e8a23"}):
  print(tag)'''
```

```python
adresse = []
for tag in soup.findAll('div', {'class':"IconFact-e8a23"}):
  if tag.i.text == "location":
    #print(tag.span.text)
    adresse.append(tag.span.text)
```

```python
besonderes = []
for tag in soup.findAll('div', {'class':"IconFact-e8a23"}):
  if tag.i.text == "check":
    #print(tag.span.text)
    besonderes.append(tag.span.text)
```

```python
df = pd.DataFrame({'Text': h2, 'Adresse':adresse, 'Preis':price, 'Quadratmeter':qm, 'Anzahl Zimmer':zimmer, 'Besonderes':besonderes})
```

```python
df
```
